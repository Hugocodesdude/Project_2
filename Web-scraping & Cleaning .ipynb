```python
def extract(something):
    str_ = ''
    nums_ = ''
    
    for i in something.getText():
        if i.isdigit() == False:
            str_ = str_ + i
        else:
            nums_ = nums_ + i
    return str_.replace(".", "").replace(",", ""), nums_
```
